```python
# 150118825 Ahmet Hakan Ekşİ
# 150117051 Metehan Ertan
# 150115025 Mert İsmail Eğİ
# natural language processing: collocations ranking
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
import pandas as pd
import os

donemIndex = os.listdir('NLP/')
print(donemIndex)

donemler= pd.DataFrame(["","" ,"","","","","","" ] ,index=donemIndex) #butun donemler burada donem.loc[donemx] diyerek erişilebilir 
```

<h1><b>Reading Transcripts</b></h1>

```python
for donem in donemIndex:
    yillar=os.listdir('NLP/'+donem)
    for yil in yillar:
        tutanaklar=os.listdir('NLP/'+donem+'/'+yil)
        for tutanakAdi in tutanaklar:
            #if tutanaklar.index(tutanakAdi)==10:
            #   break
            f = open('NLP/'+donem+'/'+yil+'/'+tutanakAdi, "r",encoding="utf8")
            tutanak = f.read()
            donemler.loc[donem]+=tutanak
    print("Tamamlanan donem",donem)
```

<h1><b>Extracting Collocations for Bigrams</b></h1>

```python
from nltk.collocations import *
import matplotlib.pyplot as plt

bigram_measures = nltk.collocations.BigramAssocMeasures()
measureIndex=["Frequency","Student T Test","Likelihood Ratio","Chi-Square Test"]
collocMethods= pd.DataFrame([bigram_measures.raw_freq,bigram_measures.student_t,bigram_measures.likelihood_ratio, bigram_measures.chi_sq]
                       ,index=measureIndex) 

for donem in donemIndex:
    if donem=="donem25":
        continue
    print("\n--------",donem.upper(),"--------")
    words = re.sub(r'[^\w\s]', '',donemler.loc[donem][0]).split()
    finder = BigramCollocationFinder.from_words(words)
    finder.apply_freq_filter(5000) #frequency filter 5000
    for collocMethod in measureIndex:
        donemColloc = []
        donemFreq = []
        print("\n",collocMethod,":")
        for i in finder.score_ngrams(collocMethods.loc[collocMethod][0]):
            if finder.score_ngrams(collocMethods.loc[collocMethod][0]).index(i)==10: #top 10 collocations for each measure
                break
            print (i)
            donemColloc.append("{" + i[0][0] + ", " + i[0][1] + "}" )
            donemFreq.append(i[1])
        plt.bar(donemColloc,donemFreq)
        plt.xticks(rotation=90)
        plt.title(donem.upper() + " : " + collocMethod)
        plt.xlabel('Collocations')
        plt.ylabel('Score')
        plt.show()      
    
```
