```python
# natural language processing: n-gram ranking
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
import pandas as pd
import os

donemIndex = os.listdir('NLP/')
print(donemIndex)

donemler= pd.DataFrame(["","" ,"","","","","","" ] ,index=donemIndex) #butun donemler burada donem.loc[donemx] diyerek erişilebilir 
```

```python
for donem in donemIndex:
    yillar=os.listdir('NLP/'+donem)
    for yil in yillar:
        tutanaklar=os.listdir('NLP/'+donem+'/'+yil)
        for tutanakAdi in tutanaklar:
            f = open('NLP/'+donem+'/'+yil+'/'+tutanakAdi, "r",encoding="utf8")
            tutanak = f.read()
            donemler.loc[donem]+=tutanak
    print("Tamamlanan donem",donem)
```

```python
for donem in donemIndex:
    print(donem,"Top 10 with stopword:")
    words = re.sub(r'[^\w\s]', '',donemler.loc[donem][0]).split()
    for i in (1,2,3):
        print(i,"-gram:")
        print (pd.Series(nltk.ngrams(words, i)).value_counts()[:10])
    
```

```python
import nltk
nltk.download('stopwords')
stopWords = set(stopwords.words('turkish'))
for donem in donemIndex:
    print(donem,"Top 10 without stopword:")
    words = re.sub(r'[^\w\s]', '', donemler.loc[donem][0]).split()
    words = [word for word in words if word not in stopWords]
    for i in (1,2,3):
        print(i,"-gram:")
        print (pd.Series(nltk.ngrams(words, i)).value_counts()[:10])
```

```python
wordsWithout=[]
wordsWith=[]
for donem in donemIndex:
    wordsWith+= re.sub(r'[^\w\s]', '', donemler.loc[donem][0]).split()
    #wordsWith+=donemler.loc[donem][0].split()
    wordsWithout += [word for word in wordsWith if word not in stopWords]

print("All corpus merged")
```

```python
for i in (1,2,3):
    print(i,"-gram for whole corpus with stopword:")
    print (pd.Series(nltk.ngrams(wordsWith, i)).value_counts()[:10])
```

```python
for i in (1,2,3):
    print(i,"-gram for whole corpus without stopword:")
    print (pd.Series(nltk.ngrams(wordsWithout, i)).value_counts()[:10])
```
